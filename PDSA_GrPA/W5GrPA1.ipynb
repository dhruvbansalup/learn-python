```python
def FiberLink(distance_map): #distance_map=[source:[destination,distance],...]
    distance_map = [(u, v, d) for u in distance_map for v, d in distance_map[u]]
    distance_map = sorted((([d[2],d[0],d[1]]) for d in distance_map), key=lambda x: x[0])

    group = {x[1]: x[1] for x in distance_map}
    group.update({x[2]: x[2] for x in distance_map})
    length = 0

    for d,u,v in distance_map:
        if group[u]!=group[v]:
            length+=d
            g=group[u]
            for i in group:
                if group[i]==g:
                    group[i]=group[v]
    
    return length
```

### Test Case 1


```python
size = 7
edges = [(0,1,10),(0,2,50),(0,3,300),(5,6,45),(2,1,30),(6,4,37),(1,6,65),(2,5,76),(1,3,40),(3,4,60),(2,4,20)]
WL = {}
for i in range(size):
    WL[i] = []
for ed in edges:
    WL[ed[0]].append((ed[1],ed[2]))
    WL[ed[1]].append((ed[0],ed[2]))
expected_output = 182
actual_output = FiberLink(WL)
print(f"Test Case | Expected - {expected_output} | Actual - {actual_output}")
```
