```python
def MaxValueSelection(items,C):
    list=[]#[[rate,item,qty],...]
    for item in items:
        list.append([(items[item][1]/items[item][0]),item,items[item][0]])
    list = sorted(list, key=lambda x:x[0], reverse=True)
    
    value=0
    count=0
    while count<C:
        currItem=list.pop(0)
        if count+currItem[2]<C:
            value+=currItem[0]*currItem[2]
            count+=currItem[2]
        else:
            value+=currItem[0]*(C-count)
            count+=C-count
    return value
```
