# Shortest Path



```python
# Prefix code
def BFSListPathLevel(AList, v):
    level, parent = {}, {}
    for i in AList.keys():
        level[i] = -1
        parent[i] = -1
    q = []

    level[v] = 0
    q.append(v)

    while len(q) > 0:
        j = q.pop(0)
        for k in AList[j]:
            if level[k] == -1:
                level[k] = level[j]+1
                parent[k] = j
                q.append(k)
    return level, parent
```

```python
#Solution
def minimumhops(AList,start,end):
    parent={x:-1 for x in AList.keys()}
    visited={x:False for x in AList.keys()}
    queue=[]
    
    queue.append(start)
    visited[start]=True
    
    while len(queue)!=0:
        x=queue.pop(0)
        for k in AList[x]:
            if not visited[k]:
                visited[k]=True
                parent[k]=x
                queue.append(k)
                if k==end:
                    break
    
    path=[]
    step=end
    while step!=-1:
        path.append(step)
        step=parent[step]
    path.reverse()

    if path[0]!=start:
        return [start]
    
    return path
```

```python
#Test Cases

start = 8
end = 7
AList = {0: [8], 8: [9], 1: [3, 5, 8], 3: [1, 7, 2], 5: [4], 2: [8, 9], 9: [1], 7: [8], 4: [2, 6], 6: [9]}
shortestpath = minimumhops(AList, start, end)
print(shortestpath)
```
