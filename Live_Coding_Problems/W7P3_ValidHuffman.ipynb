```python
def IsCodeValid(hfcode, message):
    # Reverse the hfcode dictionary to create a decoding map
    code = {v: k for k, v in hfcode.items()}
    
    decoded = ""
    currCode = ""
    
    # Iterate through each character in the message
    for i in message:
        currCode += i
        # Check if the current code exists in the decoding map
        if currCode in code:
            decoded += code[currCode]
            currCode = ''
    
    # If currCode is empty, the entire message was successfully decoded
    return currCode == ""
```
