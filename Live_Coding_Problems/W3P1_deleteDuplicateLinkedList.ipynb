```python
def removeDuplicate(head):
    pnode=head
    pdata=head.data
    
    temp=head.next
    
    while temp!=None:
        nnode=temp.next
        
        if pdata==temp.data:
            pnode.next=nnode
            temp=temp.next
            continue
        
        pnode=temp
        pdata=temp.data
        temp=temp.next
        
        
```
