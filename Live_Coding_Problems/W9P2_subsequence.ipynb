```python
def countSubseq(S):
    n = len(S)
    count = [1] * n  # Every character alone forms a valid subsequence

    # Process from the second last character to the first
    for i in range(n - 2, -1, -1):
        for j in range(i + 1, n):
            if int(S[i]) < int(S[j]):  # Valid subsequence
                count[i] += count[j]

    # Return the total count of subsequences
    return sum(count)
```
